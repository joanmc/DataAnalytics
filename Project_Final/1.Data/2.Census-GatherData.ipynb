```python
# Extract the number of under 18 Male in each area by summing each age group
ages_male = df_ages[['GEOGDESC', 'T1_1AGE0M', 'T1_1AGE1M', 'T1_1AGE2M', 'T1_1AGE3M', 'T1_1AGE4M', 'T1_1AGE5M',
                             'T1_1AGE6M', 'T1_1AGE7M', 'T1_1AGE8M', 'T1_1AGE9M', 'T1_1AGE10M', 'T1_1AGE11M','T1_1AGE12M',
                             'T1_1AGE13M', 'T1_1AGE14M', 'T1_1AGE15M', 'T1_1AGE16M', 'T1_1AGE17M', 'T1_1AGE18M']]

# Sum Males under 18
col_list = list(ages_male)
ages_male['Total_Under18Male'] = ages_male[col_list].sum(axis=1)

# Add the column containing the total poopulation of under 18 Male to the data frame containing all selected census data information
Total_U18_Male = ages_male['Total_Under18Male']
Total_U18_Male = Total_U18_Male.reset_index(drop=True)
df_stations['Total_Under18Male'] = Total_U18_Male
```

```python
# Extract the number of under 18 Female in each area
ages_female = df_ages[['GEOGDESC', 'T1_1AGE0F', 'T1_1AGE1F', 'T1_1AGE2F', 'T1_1AGE3F', 'T1_1AGE4F', 'T1_1AGE5F',
                             'T1_1AGE6F', 'T1_1AGE7F', 'T1_1AGE8F', 'T1_1AGE9F', 'T1_1AGE10F', 'T1_1AGE11F','T1_1AGE12F',
                             'T1_1AGE13F', 'T1_1AGE14F', 'T1_1AGE15F', 'T1_1AGE16F', 'T1_1AGE17F', 'T1_1AGE18F']]

# Sum Females under 18
col_list_f = list(ages_female)
ages_female['Total_Under18Female'] = ages_female[col_list_f].sum(axis=1)

# Add the column containing the total poopulation of under 18 Female to the data frame containing all selected census data information
Total_U18_Female = ages_female['Total_Under18Female']
Total_U18_Female = Total_U18_Female.reset_index(drop=True)
df_stations['Total_Under18Female'] = Total_U18_Female
```

```python
# Find the total population under 18 (by summing under 18 Male and Female population) for each subdivision
# add this total to the dataframe containing all census information
Total_U18 = [Total_U18_Male[i] + Total_U18_Female[i] for i in range(0, len(Total_U18_Male))]
df_stations['Total_Under18'] = Total_U18
df_stations.head()
```
