```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
%matplotlib inline

df = pd.read_csv('PerCapita.csv')
df = df.drop('Unnamed: 0', 1)
df.head(33)
```

```python
df.dtypes
```

```python
df.corr()
```

```python
lm = sm.ols(formula="TotalCrime ~ Green + NoLowEducation  + GoodHealth + HigherSocialClass + PopulationDensity", data=df).fit()
print(lm.params)
```

```python
print(lm.summary())
```

```python
GreenClass = (df.Green > 7.3) * 1.0
df_GreenClass = pd.DataFrame({'GreenClass': GreenClass})

CrimeClass = (df.TotalCrime > 607.625) * 1.0
df_CrimeClass = pd.DataFrame({'CrimeClass': CrimeClass})

df_GreenCrimeClass = pd.concat([df_CrimeClass, df_GreenClass], axis=1)

df_classes = pd.concat([df, df_GreenCrimeClass], axis=1)
#df_allcrimeclass.drop('TotalCrime', axis=1, inplace=True)
df_classes
```

```python
lm = sm.ols(formula="TotalCrime ~ GreenClass-1", data=df).fit()
print(lm.params)
```

```python
print(lm.summary())
```

```python
lm = sm.ols(formula="HigherSocialClass ~ Green + HighEducation", data=df).fit()
print(lm.params)
```

```python
print(lm.summary())
```
