```python
GreenClass = (df.Green > 7.3) * 1.0
df_GreenClass = pd.DataFrame({'GreenClass': GreenClass})

CrimeClass = (df.TotalCrime > 607.625) * 1.0
df_CrimeClass = pd.DataFrame({'CrimeClass': CrimeClass})

df_GreenCrimeClass = pd.concat([df_CrimeClass, df_GreenClass], axis=1)

df_classes = pd.concat([df, df_GreenCrimeClass], axis=1)
#df_allcrimeclass.drop('TotalCrime', axis=1, inplace=True)
df_classes
```

```python
df1 = pd.crosstab([df_classes.CrimeClass], df_classes.Green)
df1.plot(kind='bar', stacked=True)
```
